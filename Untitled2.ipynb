

*   Na regressão logistica utiliza os dados quantitativos, então é necessário fazer a transformação dos dados qualitativos para quantitativos


```r
#data transformation to quantitative (function mutate)
dados_quant <- dados_1
colnames(dados_quant)

dados_quant %>%
    mutate(customerID = NULL, #for this column is not necessary transformation
            PhoneService = as.factor(ifelse(PhoneService == "Yes",1,0)),
            Partner = as.factor(ifelse(Partner == "Yes",1,0)),
            Dependents = as.factor(ifelse(Dependents == "Yes",1,0)),
            PaperlessBilling = as.factor(ifelse(PaperlessBilling == "Yes",1,0)),
            Churn = as.factor(ifelse(Churn == "Yes",1,0)),
    ) -> dados_quant

  glimpse(dados_quant)
```

```r
#transformation with dummy (for more than two category)
dummy_dados <- dados_quant %>% select (InternetService, Contract, PaymentMethod,
                                        MultipleLines, OnlineBackup, OnlineSecurity,
                                        DeviceProtection, StreamingTV, StreamingMovies,
                                        TechSupport)

dummy <- dummyVars(~., data = dummy_dados, fullRank = T)

dummy_dados <- predict(dummy,dummy_dados)
dados_quant1<-bind_cols(dados_quant,dummy_dados)

#rename columns
dados_quant1 %>%
      rename( InternetService.Fiberoptic = 'InternetService.Fiber optic',
              Contract.Oneyear = 'Contract.One year',
              Contract.Twoyear = 'Contract.Two year',
              PaymentMethod.Creditcard = 'PaymentMethod.Credit card (automatic)',
              PaymentMethod.Electronic = 'PaymentMethod.Electronic check',
              PaymentMethod.Mailed = 'PaymentMethod.Mailed check',
              MultipleLines.NoService = 'MultipleLines.No phone service',
              OnlineBackup.NoService = 'OnlineBackup.No internet service',
              OnlineSecurity.NoService = 'OnlineSecurity.No internet service',
              DeviceProtection.NoService = 'DeviceProtection.No internet service',
              StreamingTV.NoService = 'StreamingTV.No internet service',
              StreamingMovies.NoService = 'StreamingMovies.No internet service',
              TechSupport.NoService = 'TechSupport.No internet service') -> dados_quant1

#delete columns qualitative
dados_quant1 %>%
  mutate(gender = NULL,
          InternetService = NULL,
          Contract = NULL,
          PaymentMethod = NULL,
          PaymentMOnlineBackupethod = NULL,
          OnlineSecurity = NULL,
          StreamingTV = NULL,
          StreamingMovies = NULL,
          MultipleLines = NULL,
          OnlineBackup = NULL,
          DeviceProtection = NULL,
          TechSupport = NULL) -> dados_quant1
glimpse(dados_quant1)
```

*Explicação dummy:


*   **dummyVars**(*~.*, data = *dummy_dados*, **fullRank** = TRUE)
    > dummyVars -> é a função para transformar em variáveis dummy
    > '~.' -> considera todas as colunas do DF
    > fullRank -> evita multicolinearidade. Se uma variável categórica tem n categorias, ela vai criar apenas n-1 dummies (eliminando uma categoria de referência> é deduzido que essa categoria é presente (1) quando é zero em todas as outras).

*   **predict**: aplica o modelo de transformação (dummy) aos dados dummy_dados. O resultado é uma matriz numérica, onde as variáveis categóricas foram convertidas em variáveis dummy.
*   **bind_cols**: junta colunas de dois data frames (ou data frame + matriz).


*Explicação do balanceamento

Validação Cruzada - avalia a capacidade de generalização do modelo

*   O modelo de validação cruzada denominado k-fold consiste em dividir o conjunto total de dados em k subconjuntos mutuamente exclusivos do mesmo tamanho e, a partir dai, um subconjunto é utilizado para teste e os k-1 restantes são utilizados para estimação dos parâmetros, fazendo-se o cálculo da acurácoa do modelo


```r
trainIndex <- createDataPartition(dados_num_balanc$Churn, p = .8,
                                  list = FALSE,
                                  times = 1)

vivoTrain <- dados_num_balanc[ trainIndex,]
vivoTest <- dados_num_balanc[-trainIndex,]
```

```r
#training
set.seed(150)

glm_model = train(Churn ~ SeniorCitizen + tenure + PaperlessBilling +
                  MonthlyCharges + TotalCharges + InternetService.Fiberoptic +
                  Contract.Oneyear + Contract.Twoyear + PaymentMethod.Electronic +
                    MultipleLines.Yes + OnlineSecurity.Yes + StreamingTV.Yes +
                    StreamingMovies.Yes + TechSupport.NoService,
                  data = vivoTrain,
                  method = 'glm',
                  trControl = trainControl(method = 'cv'),
                  family = "binomial"
                  )

summary(glm_model)
varImp(glm_model)
```

```r
#test base
reg_log_pred <- predict(glm_model,vivoTrain)

reg_log_pred1 <- data.frame(reg_log_pred=reg_log_pred)

reg_log_pred1$reg_log_pred <- as.factor(ifelse(reg_log_pred1$reg_log_pred >= 0.5,1,0))
# reg_log_pred1$reg_log_pred <- as.factor(ifelse(reg_log_pred1$reg_log_pred ==1,"evadido", "cliente"))
reg_log_pred1$reg_log_pred <- factor(reg_log_pred1$reg_log_pred, levels = c(0,1), labels = c("cliente","evadido"))

# vivoTrain$Churn <- as.factor(vivoTrain$Churn)
vivoTrain$Churn <- as.factor(ifelse(vivoTrain$Churn==1,"evadido","cliente"))
vivoTrain$Churn <- factor(vivoTrain$Churn, levels = c("cliente","evadido"))

glimpse(reg_log_pred1)

matrix_reg <- confusionMatrix(data = reg_log_pred1$reg_log_pred, reference = vivoTrain$Churn, positive = "evadido")
matrix_reg$table

metricas <- data.frame(matrix_reg$byClass)
```

```r
#view data
glimpse(dados_1)

#delete ID / transform quati to quali

dados_1 %>%
    mutate(SeniorCitizen = as.factor(ifelse(SeniorCitizen == 1, "Yes", "No")),
            customerID = NULL) -> dados_quali
```

```r
#create class (quali)
##view
tenure <- summary(dados_quali$tenure)
tenure
##create
min_tenure <- tenure[[1]]-5
q1_tenure <- tenure[[2]]
q2_tenure <- tenure[[3]]
q3_tenure <- tenure[[5]]
max_tenure <- tenure[[6]]+5

dados_quali %>%
    mutate(tenure = cut(tenure, breaks = c(min_tenure,
                                                      q1_tenure,
                                                      q2_tenure,
                                                      q3_tenure,
                                                      max_tenure))) -> dados_quali
summary(dados_quali$tenure)
```

```r
#create class (quali)
##view
MonthlyCharges <- summary(dados_quali$MonthlyCharges)

##create
min_MonthlyCharges <-MonthlyCharges[[1]]-5
q1_MonthlyCharges<- MonthlyCharges[[2]]
q2_MonthlyCharges<- MonthlyCharges[[3]]
q3_MonthlyCharges<- MonthlyCharges[[5]]
max_MonthlyCharges <-MonthlyCharges[[6]]+5

dados_quali %>%
    mutate(MonthlyCharges = cut(MonthlyCharges, breaks = c(min_MonthlyCharges,
                                                      q1_MonthlyCharges,
                                                      q2_MonthlyCharges,
                                                      q3_MonthlyCharges,
                                                      max_MonthlyCharges))) -> dados_quali
summary(dados_quali$MonthlyCharges)
```

```r
#create class (quali)
##view
TotalCharges <- summary(dados_quali$TotalCharges)

##create
min_TotalCharges <-TotalCharges[[1]]-5
q1_TotalCharges<- TotalCharges[[2]]
q2_TotalCharges<- TotalCharges[[3]]
q3_TotalCharges<- TotalCharges[[5]]
max_TotalCharges <-TotalCharges[[6]]+5

dados_quali %>%
    mutate(TotalCharges = cut(TotalCharges, breaks = c(min_TotalCharges,
                                                      q1_TotalCharges,
                                                      q2_TotalCharges,
                                                      q3_TotalCharges,
                                                      max_TotalCharges))) -> dados_quali
summary(dados_quali$TotalCharges)
```

```r
#database balancing
glimpse(dados_quali)

dados_quali %>%
    select(Churn) %>%
    group_by(Churn) %>%
    summarise(n = n())

dados_quali %>%
    filter(Churn == "No") %>%
    sample_n(1869) -> dados_quali_no

dados_quali %>%
    filter(Churn == "Yes") -> dados_quali_yes

dados_quali_balanc <- bind_rows(dados_quali_no,dados_quali_yes)

dados_quali_balanc %>%
    select(Churn) %>%
    group_by(Churn) %>%
    summarise(n = n())
```

```r
trainIndex_quali <- createDataPartition(dados_quali_balanc$Churn, p = .8,
                                  list = FALSE,
                                  times = 1)

vivoTrain_quali <- dados_quali_balanc[ trainIndex_quali,]
vivoTest_quali <- dados_quali_balanc[-trainIndex_quali,]
```
